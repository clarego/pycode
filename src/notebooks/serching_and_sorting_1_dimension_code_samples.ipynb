```python
numbers = [64, 34, 25, 12, 22, 11, 90]

# ── Bubble Sort ──────────────────────────────
n = len(numbers)

for i in range(n):
    for j in range(0, n - i - 1):
        if numbers[j] > numbers[j + 1]:
            # Swap the two adjacent elements
            temp = numbers[j]
            numbers[j] = numbers[j + 1]
            numbers[j + 1] = temp

print("Bubble sorted:", numbers)
```

```python
numbers = [64, 34, 25, 12, 22, 11, 90]

# ── Selection Sort ───────────────────────────
n = len(numbers)

for i in range(n):
    # Assume the current position holds the minimum
    min_index = i

    for j in range(i + 1, n):
        if numbers[j] < numbers[min_index]:
            min_index = j

    # Swap the found minimum with the current position
    temp = numbers[i]
    numbers[i] = numbers[min_index]
    numbers[min_index] = temp

print("Selection sorted:", numbers)
```

```python
numbers = [64, 34, 25, 12, 22, 11, 90]

# ── Insertion Sort ───────────────────────────
n = len(numbers)

for i in range(1, n):
    key = numbers[i]      # element to be inserted
    j = i - 1

    # Shift elements greater than key one position right
    while j >= 0 and numbers[j] > key:
        numbers[j + 1] = numbers[j]
        j -= 1

    numbers[j + 1] = key  # place key in correct spot

print("Insertion sorted:", numbers)
```

```python
numbers = [64, 34, 25, 12, 22, 11, 90]
# ── Quick Sort ───────────────────────────────
def quick_sort(arr, low, high):
    if low < high:
        pivot_index = partition(arr, low, high)
        quick_sort(arr, low, pivot_index - 1)
        quick_sort(arr, pivot_index + 1, high)

def partition(arr, low, high):
    pivot = arr[high]     # use last element as pivot
    i = low - 1          # pointer for smaller elements

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            # Swap arr[i] and arr[j]
            temp = arr[i]
            arr[i] = arr[j]
            arr[j] = temp

    # Place pivot in its correct position
    temp = arr[i + 1]
    arr[i + 1] = arr[high]
    arr[high] = temp
    return i + 1

quick_sort(numbers, 0, len(numbers) - 1)
print("Quick sorted:", numbers)
```

```python
numbers = [64, 34, 25, 12, 22, 11, 90]
target = 22

# ── Linear Search ────────────────────────────
found_index = -1

for i in range(len(numbers)):
    if numbers[i] == target:
        found_index = i
        break          # stop as soon as we find it

if found_index != -1:
    print(f"Found {target} at index {found_index}")
else:
    print(f"{target} not found in list")
```

```python
# Binary search needs a SORTED list first
numbers = [11, 12, 22, 25, 34, 64, 90]
target = 25

# ── Binary Search ────────────────────────────
low = 0
high = len(numbers) - 1
found_index = -1

while low <= high:
    mid = (low + high) // 2   # find the middle index

    if numbers[mid] == target:
        found_index = mid
        break
    elif numbers[mid] < target:
        low = mid + 1          # target is in right half
    else:
        high = mid - 1         # target is in left half

if found_index != -1:
    print(f"Found {target} at index {found_index}")
else:
    print(f"{target} not found in list")
```
